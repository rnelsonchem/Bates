# Simulation and Stack Plotting

In this demonstration, we'll make a very simple plot of a couple of simulated mathematical functions, and then we will use this to illustrate how you would be able to make a stack plot of spectra like the [one shown below](http://www.mathworks.com/matlabcentral/fileexchange/24368-stacked-plot).

<img src="static/stack_plot.jpg">

To start, let's consider what it takes to make a plot of a mathematical function (or some data you've collected). A plot in a [Cartesian coordinate system](http://en.wikipedia.org/wiki/Cartesian_coordinate_system) is generated by taking at the minimum two points -- $(x_1, y_1)$ and $(x_2, y_2)$ -- and connecting them with a line. This is shown on the left in the figure below. If we have another point, $(x_3, y_3)$, we can add that point to our plot and draw a line connecting this new point to the previous one (center). If we continue this process for a large number of points, we would add all of these points to our graph and connect them with a line (right).

<img src="static/Plot basics.svg">

Using this procedure, it is fairly simple to see how you would plot a data set that you collect in the lab -- e.g. concentration (y data) versus time (x dat) How can we use this procedure to simulate a mathematical function? In a cartesian coordiate system, we have to remember that the value of any mathematical function ($f$) will be related to some input value. If we consider our input to the function to be our x data, then we can define the output of our function as the y data. 

$$ y_n = f(x_n) $$

If we wanted to plot this function, we should fisrt pick some values of x and determine the value of the function $f$ for each of those points:

$$ y_1 = f(x_1) $$
$$ y_2 = f(x_2) $$
$$ \ldots $$
$$ y_n = f(x_n) $$

Or if we consider all the x values simultaneously as an *array*, the we can generalize this expression:

$$ [y_1, y_2, ... , y_n] = f([x_1, x_2, ... , x_n])$$

## Let's write some code

A common convention in Python is to load any third party code at the beginning of any program using the `import` statement. In this case, we'll load up two external cody libraries: [Numpy](http://www.numpy.org/) makes it easy to work with numbers in Python, and [Matplotlib](http://matplotlib.org/) is a very nice plotting library. We'll use the `as` notation to give these libraries a nickname. That way we don't have to type out their full name every time we want to use them.

```python
import numpy as np
import matplotlib.pyplot as plt
```

First of all, let's decide on a range of x values to use for our simulation. In this case, we'll use the [Numpy function `linspace`](http://docs.scipy.org/doc/numpy/reference/generated/numpy.linspace.html) to generate a *lin*ear *space*d array of numbers. The first value in the parentheses is the starting value, the last value in the end, and the third value is the total number of points in our array. We can print the result to see what it looks like.

```python
x = np.linspace(0, 2*3.14, 5)
print x
```

To generate our y data, we need to pick a mathematical function that we want to simulate, and then we'll determine the value of that function for every x value. In this case, let's try out a [sine function](http://en.wikipedia.org/wiki/Sine). Numpy as a [`sin` function](http://docs.scipy.org/doc/numpy/reference/generated/numpy.sin.html) that is nice because we can pass in our array of x values and it will automatically give us all the corresponding sine values.

```python
y = np.sin(x)
print y
```

We can use [Matplotlib's `plot` function](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot) to visualize this data. In this case, I want to plot my x vs y data using circular markers connected by lines (`'o-'`).

```python
plt.plot(x, y, 'o-')
```

This looks more like a saw-tooth function than a sine function. That's because we only used 5 data points. Let's replot this with more points to see how it looks. Change the last number in the `linspace` function to see what it looks like with different numbers of points.

```python
x = np.linspace(0, 2*3.14, 10)
y = np.sin(x)

plt.plot(x, y, 'o-')
```

The more data points we use in our x array, the more faithfully our y data mimics the sine function. For fun, we can do this with other functions as well to compare them. In this case, let's use lots of x data points, so that our plots look very smooth.

```python
x = np.linspace(0, 2*3.14, 100000)

y1 = np.sin(x)
y2 = np.sin(2*x)
```

For our plot, we have so many points that it doesn't make sense to use the circular markers anymore.

```python
plt.plot(x, y1)
plt.plot(x, y2)
```

## Stack plotting

Okay. Now that we've learned how to simulate a mathematical function, let's see if we can think about how to make a stack plot like we saw in the beginning. Take a look at that picture and think a little about what is happening. If we plotted that data normally, everything would overlap in the same way that our functions overlap in the simulation above. What is different between the first data set line (lowest one) and the second data set line? Well really it looks like the x data starts and ends at a slightly larger value and the lowest point (baseline) of the y data is at slightly larger overall values. This analysis gives us an idea of how we might accomplish this task, and it's really quite simple. Let's just add a small constant value to both the x and y data. That causes an *offset* of the second data set relative to our first one.

```python
plt.plot(x, y1)
plt.plot(x + 0.5, y2 + 0.5)
```

Great! Now you've learned how to make a stack plot. If you have a bunch of data that you'd like to plot in this way, you could simply repeat this process for every data set.

## Putting it all together

Below is all of the minimal code from above that you would need to recreate this example.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*3.14, 100000)

y1 = np.sin(x)
y2 = np.sin(2*x)

plt.plot(x, y1)
plt.plot(x + 0.5, y2 + 0.5)
plt.show() # This is necessary if you want to run this outside the notebook
```
